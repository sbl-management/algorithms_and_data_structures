-   Paraphrase the problem in your own words.


The task is to identify numbers missing from a given list that should contain values ranging from 0 to n. Some numbers might not be present, and the objective is to return a list of these missing values. If every number in the range is accounted for, the function should return -1. The list may include duplicate values, but they do not impact the outcome.


-   Create 1 new example that demonstrates you understand the problem. Trace/walkthrough 1 example that your partner made and explain it.


Input: lst = [4, 3, 1, 0, 6]
Output: [2, 5]
The full range should be [0, 1, 2, 3, 4, 5, 6]. The missing numbers in the list are 2 and 5.


-   Copy the solution your partner wrote. 


```python
from typing import List, Union

def missing_num(nums: List[int]) -> Union[int, List[int]]:
    n = max(nums)
    
    # baseline of numbers
    baseline_nums = set(range(n + 1))

    # input into set
    nums_set = set(nums)
    
    # checking the difference 
    missing = list(baseline_nums - nums_set)
    
    # If there are no missing numbers, return -1
    if not missing:
        return -1
    else: 
        return missing

print(missing_num(nums = [0, 2]))  # Output: [1]
print(missing_num(nums = [5, 0, 1]))  # Output: [2, 3, 4]
print(missing_num(nums = [6, 8, 2, 3, 5, 7, 0, 1, 10]))  # Output: [4, 9]
```


-   Explain why their solution works in your own words.


The solution starts by determining the maximum number, n, in the list, which sets the full range from 0 to n. It then creates a set with all the numbers in this range and another set from the input list to remove any duplicates. The missing numbers are found by calculating the difference between these two sets. If no missing numbers exist, the function returns -1. This method efficiently identifies missing numbers and accounts for any duplicates in the input.


-   Explain the problem’s time and space complexity in your own words.


The time complexity of this solution is O(n), where n is the number of elements in the list. This is because we are creating sets, which require iterating through the list to convert it. The set operations (union and difference) also take linear time.

The space complexity is also O(n) because we are storing two sets—one for the full range and another for the input list, each requiring space proportional to the number of elements.


-   Critique your partner's solution, including explanation, and if there is anything that should be adjusted.


The solution is straightforward and works effectively. Using sets to find the missing numbers is a good approach as it handles duplicates and makes the difference operation simpler. However, there’s a potential issue with relying on the maximum value in the list to define the range. If a number greater than n exists, it could cause inaccuracies. A safer approach would be to use the length of the list (len(nums)) to determine the range. Aside from this, the solution is solid and handles the problem efficiently.
